Q1. What is the average, median, and standard deviation of house prices?

```python
df['Price'].describe()
```

```python
average_price = df['Price'].mean()
median_price = df['Price'].median()
Sd_price = df['Price'].std()

print(average_price)
print(median_price)
print(Sd_price)
```

Explanation

mean() calculates the average house price.

median() finds the middle value when prices are arranged in order.

std() calculates the standard deviation, which measures how spread out the house prices are.

Interpretation of Output

The average house price is ₹540,088.

The median price is ₹450,000, which is lower than the mean, suggesting that some expensive houses increase the average price.

The standard deviation of ₹367,127 indicates a wide variation in house prices across the dataset.

This suggests that the dataset contains both low-priced and high-priced houses, leading to a broad distribution of prices.

Q3. Check if the price data is skewed.


```python
print(df['Price'].skew())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Price'], kde=True)
plt.title("Distribution of House Prices")
plt.show()
```

Explanation

The .skew() function measures the asymmetry of the data distribution.

A skewness value close to 0 indicates a symmetric distribution.

A positive skewness value means the data is right-skewed.

A negative skewness value means the data is left-skewed.

Interpretation

The skewness value of the Price column is 4.27, which is much greater than 1. Therefore, the price data is highly positively skewed (right-skewed). This means that most houses have relatively lower prices, while a few houses have very high prices, creating a long tail on the right side of the distribution.

Q4. What is the average price for each number of bedrooms?


```python
avg_price_bedroom = df.groupby('number of bedrooms')['Price'].mean()
print(avg_price_bedroom)
```

Explanation

groupby('number of bedrooms') groups houses based on the number of bedrooms.
['Price'].mean() calculates the average price for each bedroom category.
Interpretation

The analysis shows that the average house price generally increases as the number of bedrooms increases. For example, houses with 1 bedroom have an average price of approximately ₹308,964, while houses with 8 bedrooms have an average price of about ₹1,208,455.

However, some bedroom categories such as 9, 10, 11, and 33 bedrooms do not follow this trend. This may be because these categories contain very few houses, causing the average price to be less reliable. The presence of a 33-bedroom house is unusual and may represent an outlier or data entry error.

Q5.  What is the relationship between living area and price? (Use : Area without Basement).


```python
rel = df['Area of the house(excluding basement)'].corr(df['Price'])
print(rel)
```

```python
sns.scatterplot(data=df, x="Area of the house(excluding basement)", y= "Price")
plt.show()
```

Explanation

The Pearson correlation coefficient is used to measure the strength and direction of the linear relationship between two numerical variables. Its value ranges from -1 to +1.

A value close to +1 indicates a strong positive relationship.

A value close to -1 indicates a strong negative relationship.

A value close to 0 indicates little or no linear relationship.
Interpretation

The correlation coefficient between Area of the house (excluding basement) and Price is 0.615. This indicates a moderate positive relationship between the two variables.

This means that houses with a larger area (excluding the basement) generally tend to have higher prices. However, since the correlation is not very close to 1, other factors such as location, number of bathrooms, house grade, and waterfront presence may also influence house prices.

Q6. Identify any anomalies where houses have high prices but low areas.


```python
avg_price = df['Price'].mean()
avg_area = df['Area of the house(excluding basement)'].mean()

anomalies = df[
    (df['Price'] > avg_price) &
    (df['Area of the house(excluding basement)'] < avg_area)
]

print(anomalies[['Price', 'Area of the house(excluding basement)']])
```

Explanation

Houses with prices higher than the average house price are considered expensive.
Houses with areas smaller than the average area are considered relatively small.
Records satisfying both conditions are potential anomalies because they have unusually high prices despite having smaller living areas.

Interpretation

The identified houses have prices above the dataset average while their house area (excluding basement) is below the average area. These properties may be considered anomalies because house prices generally increase with area. Such cases may occur due to premium locations, waterfront views, better house grades, recent renovations, or other factors that significantly increase property value.
