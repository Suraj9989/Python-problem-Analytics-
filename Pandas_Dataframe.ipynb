Q1. Creating a DataFrame
Create a Pandas DataFrame using a Python dictionary with the following columns:
Product
Region
Sales
Quantity

```python
import pandas as pd
data = {
    "Product" : ["Phone", "TV", "Fan", "Moniter", "Charger"],
    "Region": ["West", "West", "South", "North", "East"],
    "Sales": [2000000, 9000000, 3500000, 2500000, 100000],
    "Quantity": [34500, 55000, 160000, 70000, 97000]
    }
df = pd.DataFrame(data)
df
```
