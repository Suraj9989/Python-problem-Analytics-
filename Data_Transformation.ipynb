Data Transformation

```python
import pandas as pd

from sklearn.preprocessing import (
    LabelEncoder,
    OneHotEncoder,
    MinMaxScaler,
    StandardScaler
)

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("/content/employee_productivity_dataset.csv")
df.head()
```

Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Salary'].fillna(df['Salary'].mean(), inplace=True)

df['Hours_Worked_Per_Week'].fillna(
    df['Hours_Worked_Per_Week'].median(),
    inplace=True
)

df['Performance_Score'].fillna(
    df['Performance_Score'].mean(),
    inplace=True
)

df.head()
```

Q2. Label Encoding

Convert these columns into numeric values using Label
Encoding:

Gender

Department

Show the updated columns.

```python
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

df['Department'] = LabelEncoder().fit_transform(df['Department'])

df[['Gender', 'Department']]
```

Q3. One-Hot Encoding

Apply One-Hot Encoding on:

Work_Mode

Location

Display the dataset and check how many new columns are created

```python
df_encoded = pd.get_dummies(
    df,
    columns=['Work_Mode', 'Location'],
    drop_first=False
)

print(df_encoded)

print("New Columns Created:",
      len(df_encoded.columns) - len(df.columns))
```

Q4. Normalization (Min-Max Scaling)

Normalize the following columns:

Salary

Hours_Worked_Per_Week

Use MinMaxScaler and display results

```python
scaler = MinMaxScaler()

df[['Salary_Normalized',
    'Hours_Normalized']] = scaler.fit_transform(
        df[['Salary',
            'Hours_Worked_Per_Week']]
)

print(df[['Salary',
          'Salary_Normalized',
          'Hours_Worked_Per_Week',
          'Hours_Normalized']])
```

Q5. Standardization (Scaling)

Apply StandardScaler on:

Age

Projects_Completed

Display the transformed values.

```python
std_scaler = StandardScaler()

df[['Age_Standardized',
    'Projects_Standardized']] = std_scaler.fit_transform(
        df[['Age',
            'Projects_Completed']]
)

print(df[['Age_Standardized',
          'Projects_Standardized']])
```

Q6. Compare Scaling Methods

Apply both:

MinMaxScaler

StandardScaler

On the Salary column.

Show both results side by side.


```python
salary_minmax = MinMaxScaler()
salary_standard = StandardScaler()

df['Salary_MinMax'] = salary_minmax.fit_transform(
    df[['Salary']]
)

df['Salary_Standard'] = salary_standard.fit_transform(
    df[['Salary']]
)

print(df[['Salary',
          'Salary_MinMax',
          'Salary_Standard']])
```

Q7. Build Preprocessing Pipeline

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline


```python
categorical_cols = [
    'Gender',
    'Department',
    'Work_Mode',
    'Location'
]

numerical_cols = [
    'Age',
    'Salary',
    'Hours_Worked_Per_Week',
    'Projects_Completed',
    'Performance_Score'
]
```

```python
preprocessor = ColumnTransformer(

    transformers=[

        (
            'cat',
            OneHotEncoder(handle_unknown='ignore'),
            categorical_cols
        ),

        (
            'num',
            StandardScaler(),
            numerical_cols
        )

    ]
)
```

```python
pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor)
    ]
)

print(pipeline)
```

Q8. Apply Pipeline

Apply the pipeline on the dataset.

Display:

Transformed dataset

Shape of final dataset


```python
X_transformed = pipeline.fit_transform(df)

print(X_transformed)

print("Shape of Final Dataset:",
      X_transformed.shape)
```

Q9. Conceptual Question

Why is scaling important in Python?


Scaling is important because different features can have different ranges.

Example:

Age = 25

Salary = 50000

Without scaling, Salary dominates Age during model training.

Benefits:

Faster model convergence.

Better model accuracy.

Prevents large-valued features from dominating.

Essential for KNN, SVM, K-Means, PCA, Neural Networks.

Q10. Conceptual Question

Why do we convert categorical data into numerical form?

Machine Learning algorithms perform mathematical  
calculations and cannot directly understand text values.

Example:

Gender

Male

Female

Algorithms cannot process these strings.

After encoding:

Male = 1

Female = 0

Benefits:

Makes data understandable to ML models.

Allows mathematical operations.

Improves model training and prediction.
