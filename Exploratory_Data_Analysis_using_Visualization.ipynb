```python
# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin because it has too many missing values
df.drop('Cabin', axis=1, inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Q1. Create a countplot to show the number of passengers by gender.

```python
sns.countplot(x='Sex', data=df)
plt.title('Number of Passengers by Gender')
plt.show()
```

Interpretation for this visualization:

The countplot shows that there were significantly more male passengers (around 577) than female passengers (around 314) on the Titanic. This indicates that males made up the majority of the passengers in the dataset.

You can also mention the exact numbers in your report:

Male passengers: 577

Female passengers: 314

Insight:

Since the number of male passengers is much higher than female passengers, any analysis involving survival rates or passenger demographics should consider this imbalance in gender distribution.

Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.

```python
class_counts = df['Pclass'].value_counts().sort_index()

class_counts.plot(
    kind='bar',
    color=['skyblue', 'orange', 'green']
)

plt.title('Number of Passengers in Each Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

Interpretation

The bar chart shows that Passenger Class 3 had the highest number of passengers (491), followed by Class 1 (216). Class 2 had the lowest number of passengers (184). This indicates that most passengers on the Titanic were traveling in third class.

Final Insight

More than half of the passengers belonged to the third class, suggesting that the Titanic primarily served lower-fare passengers compared to first- and second-class passengers.

Q3. Create a histogram to show the distribution of passenger age.

```python
sns.histplot(df['Age'], bins=20, color='skyblue')

plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

plt.show()
```

Interpretation

The histogram shows that most passengers were between 20 and 40 years old. There were fewer children and elderly passengers on the Titanic.

Insight

The passenger age distribution is concentrated around young and middle-aged adults, with relatively few passengers above 60 years of age.

Q4. Create a boxplot to detect outliers in the fare column.

```python
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.xlabel('Fare')
plt.show()
```

Interpretation: Several outliers are present in the Fare column.

Q5. Create a scatterplot to

a. show the relationship between age and fare.

b. analyze the relationship between family size (SibSp + Parch) and fare.

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']

sns.scatterplot(x='FamilySize', y='Fare', data=df)
plt.title('Family Size vs Fare')
plt.xlabel('Family Size')
plt.ylabel('Fare')
plt.show()
```

Q5(a). Age vs Fare

Interpretation: There is no strong relationship between age and fare.

Insight: Ticket prices depended more on passenger class than on the passenger's age.

Q5(b). Family Size vs Fare

Interpretation: Larger families generally paid higher fares.

Insight: Passengers traveling with family often spent more on tickets, possibly because they booked multiple tickets together or traveled in better classes.

Q6. Create a countplot to show survival count by gender.


```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

Interpretation: Female passengers had a much higher survival rate than male passengers.

Insight: The evacuation process appears to have prioritized women, supporting the "women and children first" policy.

Q7. Create a bar chart to show survival by passenger class.


```python
survival_rate = df.groupby('Pclass')['Survived'].mean().reset_index()

sns.barplot(
    x='Pclass',
    y='Survived',
    data=survival_rate,
    hue='Pclass',
    palette='Set2',
    legend=False
)

plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

Interpretation: First-class passengers had the highest survival rate.

Insight: Wealthier passengers had a better chance of survival than passengers in lower classes.

Q8. Create a heatmap to show correlation between numerical variables.


```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```

Interpretation: Fare and survival are positively correlated, while Pclass and Fare are negatively correlated.

Insight: Higher-paying passengers were more likely to survive, and first-class passengers generally paid higher fares.

Q9. Identify which passenger class paid the highest fare on average.


```python
avg_fare = df.groupby('Pclass')['Fare'].mean().reset_index()

sns.barplot(
    x='Pclass',
    y='Fare',
    data=avg_fare,
    hue='Pclass',
    palette='Set2',
    legend=False
)

plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()

print(avg_fare)
```

Interpretation: First-class passengers paid the highest average fare.

Insight: Ticket prices clearly reflected the social and economic differences between passenger classes.
