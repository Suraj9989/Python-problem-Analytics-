`Q7 Pivot Table Creation`
Create a pivot table showing:
1. Rows → Region
2. Columns → Product
3. Values → Total Sales
##### Which region sells the most laptops?

```python
pivot_table = df.pivot_table(
    values="Sales",
    index="Region",
    columns="Product",
    aggfunc="sum"
)
print(pivot_table["Laptop"].idxmax(), "Region sells the most laptop")
pivot_table
```
