Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the
dataset information to understand its structure.

```python
import pandas as pd
df = pd.read_csv("/content/retail_dataset .csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
print(df.isna().sum())
print(df.isna().count())
```

```python
null_values = df.isnull()
null_values
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Price"] = df["Price"].fillna(df["Price"].mean())
```

```python
print(df['Quantity'].isnull().sum())
print(df['Price'].isnull().sum())
```

 Remove all rows where the Product Category or Region column contains missing values

```python
df.dropna(subset=['Product Category', 'Region'], inplace=True)
print(df[['Product Category', 'Region']].isnull().sum())
```
